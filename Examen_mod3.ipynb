# FASE 1 (PARTE II)

- Transformación de datos

- Nulos Columna Salario 

```python
#Se transformaron los salarios negativos a positivos porque la cifra es coherente con el mercado, se determina que el signo es un error . 
df_loyalty["Salary"] = df_loyalty["Salary"].abs() #abs (valor absoluto +)
```

```python
#compruebo hacia donde se inclinan los salarios
plt.figure(figsize=(8, 5))
sns.boxplot(x=df_loyalty["Salary"], color="skyblue")
plt.title("Distribución de Salarios y Outliers")
plt.show()
```

```python
#Los nulos estan en la clasificación collage
conteo_nulos = df_loyalty[df_loyalty["Salary"].isna()].groupby("Education").size()
conteo_nulos
```

```python
# He decidido amputar por la mediana de Collage
mediana_college = df_loyalty[df_loyalty["Education"] == "College"]["Salary"].median()
```

```python
df_loyalty["Salary"] = df_loyalty["Salary"].fillna(mediana_college)
```

```python
df_loyalty["Salary"].isna().sum()
```

```python
#Unificamos los archivos
df_unificado = pd.merge(df_flight, df_loyalty, on="Loyalty Number", how="left")
```

```python
#verifico nuevas dimensiones
df_unificado.shape
```

```python
#compruebo el cambio
df_unificado.describe(include=np.number).T
```
