# FASE 1 (PARTE II)

- Transformación de datos

- Nulos Columna Salario 

```python
#Se transformaron los salarios negativos a positivos porque la cifra es coherente con el mercado, se determina que el signo es un error . 
df_loyalty["Salary"] = df_loyalty["Salary"].abs() #abs (valor absoluto +)
```

```python
#Los nulos estan en la clasificación collage
conteo_nulos = df_loyalty[df_loyalty["Salary"].isna()].groupby("Education").size()
conteo_nulos
```

```python
#no puedo amputar por a mediana de collage porque es nan completa
df_loyalty.groupby("Education")["Salary"].median()
```

```python
#compruebo hacia donde se inclinan los salarios
plt.figure(figsize=(8, 5))
sns.boxplot(x=df_loyalty["Salary"], color="skyblue")
plt.title("Distribución de Salarios y Outliers")
plt.show()
```

```python
#he decidido amputar por la mediana global
df_loyalty["Salary"] = df_loyalty["Salary"].fillna(df_loyalty["Salary"].median())
```

```python
df_loyalty["Salary"].isna().sum()
```

```python
#Unificamos los archivos
df_unificado = pd.merge(df_flight, df_loyalty, on="Loyalty Number", how="left")
```

```python
#verifico nuevas dimensiones
df_unificado.shape
```

```python
#compruebo el cambio
df_unificado.describe(include=np.number).T
```

 - Columnas Cancellation Year y Cancellation Month: He decidido mantener los nulos en las fechas de calcelación . Es lógico que existan, indican que el cliente no ha cancelado su cuenta y cambiarlos como errores sería incorrecto

```python
# Creo una nueva columna (Status) donde me indique si la membresia esta activa o inactiva, es mas funcional.
df_unificado["Status"] = np.where(df_unificado["Cancellation Year"].isnull(), "Activo", "Inactivo") #np.where Si el valor de la columna es nulo transformalo a activo sino está inactivo 
df_unificado.head(2)
```

```python
#verifico cantidades en columna Status
df_unificado["Status"].value_counts()
```

# FASE 2
- Análisis Estadístico

```python
# Estadistica descriptiva Variables numericas
df_unificado.describe(include=np.number).T
```

```python
#Parece que no depende del sueldo la cantidad de viajes 

bins = [0, 5, 15, 32]
labels = ['0-5 (Bajo)', '6-15 (Medio)', '16-32 (Alto)']
df_unificado['Flight_Range'] = pd.cut(df_unificado['Total Flights'], bins=bins, labels=labels)


plt.figure(figsize=(10, 6))
sns.boxplot(x='Flight_Range', y='Salary', data=df_unificado, palette='viridis')

plt.title('¿Ganan más los que más vuelan?', fontsize=16, fontweight='bold')
plt.xlabel('Rango de Vuelos Totales', fontsize=13)
plt.ylabel('Salario Anual', fontsize=13)
plt.show()
```

```python
#Casi Nadie viaja acompañado
sns.histplot(df_unificado["Flights with Companions"], color='skyblue', discrete=True)

plt.title('¿Cuánta gente viaja sola vs. acompañada?')
plt.xlabel('Número de Acompañantes')
plt.ylabel('Cantidad de Vuelos')
plt.xticks(range(0, 12))
plt.show()
```

```python
#la mayoria no está canjeando los puntos, quizas la cantidad de puntos para desbloquear un premio debe ser alto para poder gastarlo.
sns.histplot(df_unificado["Points Accumulated"], kde=True, color='gold', bins=50)
plt.axvline(df_unificado["Points Accumulated"].mean(), color='red', linestyle='--', label='Media')
plt.axvline(df_unificado["Points Accumulated"].median(), color='green', linestyle='-', label='Mediana')

plt.title('Distribución de Puntos Acumulados: ¿Cuánto ahorran nuestros clientes?')
plt.xlabel('Puntos Acumulados')
plt.ylabel('Cantidad de Registros')
plt.legend()
plt.show()
```

-Análisis de variables categóricas:

- Distribución de frecuencias de las variables categóricas relevantes.

```python
# Estadistica descriptiva Variables numericas
df_unificado.describe(include="str").T
```

```python
# Los grupos de clientes que viven en las provincias de Ontario, British Columbia y Quebec son el grupo mas grande
sns.set_theme(style="whitegrid")
plt.figure(figsize=(12, 6))

orden = df_unificado['Province'].value_counts().index
sns.countplot(data=df_unificado, y='Province', order=orden, palette='viridis')
plt.title('Distribución de Clientes por Provincia', fontsize=15, fontweight='bold')
plt.xlabel('Número de Clientes')
plt.ylabel('Provincia')
plt.show()
```

```python
#clasificación de clientes por segmento de categoria 

data = df_unificado['Loyalty Card'].value_counts()

plt.figure(figsize=(8, 8))
plt.pie(data, labels=data.index, autopct='%1.1f%%', startangle=140, 
        colors=['#4e79a7', '#f28e2b', '#e15759'], wedgeprops={'edgecolor': 'white'})

centre_circle = plt.Circle((0,0), 0.70, fc='white')
fig = plt.gcf()
fig.gca().add_artist(centre_circle)

plt.title('Cuota de Mercado por Tipo de Tarjeta', fontsize=15, fontweight='bold')
plt.show()
```

```python
#Los solteros Viajan un  poco mas que los casados y divorciados
sns.barplot(data=df_unificado, x='Marital Status', y='Total Flights', palette='viridis', ci=None)

plt.title('Promedio de Vuelos por Estado Civil', fontsize=15, fontweight='bold')
plt.xlabel('Estado Civil')
plt.ylabel('Promedio de Vuelos Reservados')
plt.show()
```
